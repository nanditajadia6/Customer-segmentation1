```python
df = pd.read_csv(r"C:\Users\nandita\MINI project\CSV Files\Hand Gestures\hand at rest.csv", header=0, skiprows=[1])

def zero_crossings(x):
    return np.where(np.diff(np.sign(x)))[0]

# plot the EMG signals for each channel
start_col = 1 # start at the second column (index 1)
end_col = 9

fig, ax = plt.subplots(8, 1, figsize=(20, 40), sharex=True)
for i, col in enumerate(df.columns[start_col:end_col]):
    ax[i].plot(df[col])
    ax[i].set_ylabel(col)

    # find the positive zero crossing points and plot them as red dots
    zero_crossings_col = zero_crossings(df[col])
    ax[i].plot(zero_crossings_col, df[col][zero_crossings_col], 'ro')

# set the x-axis label and show the plot
ax[7].set_xlabel('Sample')
plt.show()
```
