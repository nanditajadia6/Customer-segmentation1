```python
import numpy as np
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Load EMG signal from CSV file
data = pd.read_csv('emg_signal.csv')
data_length = len(data)
segment_length = 5000 # 500 data points per segment
segments = []
for i in range(0, data_length - segment_length, segment_length):
    segment = data[i:i+segment_length]
    segments.append(segment)
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
for i, segment in enumerate(segments):
    plt.plot(segment + i*segment_length, 'k')
plt.xlabel('Sample number')
plt.ylabel('EMG amplitude')
plt.show()
```
